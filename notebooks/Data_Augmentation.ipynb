# Data Augmentation For Classification Network

## Imports

```python
import PIL.Image as Image
import torch
import torchvision.transforms as transforms
```

## Sample Transformations

```python
base_image = Image.open("../pipeline_data_tcia/normal/normal.png")
aux = transforms.Resize((614,499))
base_image = aux(base_image)
base_image
```

### Horizontal Flip

```python
horizontal_flip = transforms.RandomHorizontalFlip(1)
horizontal = horizontal_flip(base_image)
horizontal
```

### Vertical Flip

```python
vertical_flip = transforms.RandomVerticalFlip(1)
vertical = vertical_flip(base_image)
vertical
```

### Sharpness

```python
sharp = transforms.RandomAdjustSharpness(2)
sharped = sharp(base_image)
sharped
```

### Color Jitter

```python
jitter = transforms.ColorJitter(brightness=[0.5,0.99], hue=[0.3,0.5], contrast=[0.5,0.99], saturation=[0.5,0.99])
jitted = jitter(base_image)
jitted
```

### Gaussian Blur

```python
gb = transforms.GaussianBlur(kernel_size=(5, 5))
blurred = gb(base_image)
blurred
```

### Prespective

```python
pres = transforms.RandomPerspective()
pected = pres(base_image)
pected
```

### AutoContrast

```python
cont = transforms.RandomAutocontrast()
contrasted = cont(base_image)
contrasted
```
